```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0, 9, 1)
A = X**3 + 3*X**2 - 54

plt.figure()
plt.scatter(3,0 , color = 'red')
plt.plot(X, A)
plt.annotate('Acidez Nula' , (3.4 , -40))
plt.xlabel('X')
plt.ylabel('A')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(-10, 11, 1)
F = abs(X-2) + abs(2*X+1) - X - 6

plt.figure()
plt.scatter(4,1 , color = 'green')
plt.plot(X, F)
plt.xlabel('X')
plt.ylabel('F')
plt.grid()
plt.show()
```

```python
X = np.arange(-2, 2.01, 0.01)
F = X**2 - np.sin(0.784*X**2) - 2

plt.figure()
plt.scatter(1.67,0, color = 'black')
plt.scatter(-1.67,0, color = 'black')
plt.plot(X, F)
plt.xlabel('X')
plt.ylabel('F')
plt.grid()
plt.show()
```
